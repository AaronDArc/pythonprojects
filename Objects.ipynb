```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        pass


class Library:
    def __init__(self, name):
        self.name = name
        self.books = list()
        pass

    def add_book(self, book):
        self.books.append(book)
        pass

    def remove_book(self, book):
        newlist = list()
        for currentbook in self.books:
            if currentbook.title != book.title or currentbook.author != book.author: #this "in" is a little more flexible than ==
                newlist.append(currentbook)
        self.books = newlist        
        pass

    def search_books(self, search_string):
        search_string = search_string.lower()
        newlist = list()
        for currentbook in self.books:
            if search_string in currentbook.title.lower() or search_string in currentbook.author.lower() :
                newlist.append(currentbook)
        return newlist
```

```python
# you can make certain class objects private to prefacing it with __, but this is not a true private, just a convention to say "don't touch this"

class Wizard:
    def __init__(self, name, stamina, intelligence):
        self.name = name
        self.__stamina = stamina
        self.__intelligence = intelligence
        self.mana = self.__intelligence * 10
        self.health = self.__stamina * 100

    def cast_fireball(self, target, fireball_cost, fireball_damage):
        if self.mana < fireball_cost:
            raise Exception(f"{self.name} cannot cast fireball")
        self.mana -= fireball_cost
        target.get_fireballed(fireball_damage)
        pass

    def is_alive(self):
        if self.health > 0:
            return True
        return False
        

    def get_fireballed(self, fireball_damage):
        fireball_damage -= self.__stamina
        self.health -= fireball_damage

    def drink_mana_potion(self, potion_mana):
        potion_mana += self.__intelligence
        self.mana += potion_mana



# meant for organization, not for security. called encapsulation.
```
